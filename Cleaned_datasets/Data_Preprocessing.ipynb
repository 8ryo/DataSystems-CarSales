```python
import pandas as pd 
```

```python
CarData = pd.read_csv('CleaningData.csv')
```

```python
CarData.head()
```

```python
print(CarData.dtypes)
```

```python
CarData = CarData.rename(columns={'Unnamed: 0': "VehicleID"})
```

```python
CarData.head()
```

```python
CarData.info()
```

```python
CarData = CarData.dropna()
```

```python
CarData.info()
```

```python
CarData['Year'] = CarData['Year'].astype('Int32')
```

```python
charactersToFilter = '-'
```

#### Filtering Rows based on Kilometres Driven

```python
filteredRows1 = CarData[CarData['Kilometres'].str.contains(charactersToFilter)]
```

```python
CarData = CarData.drop(filteredRows1.index)
```

```python
CarData.info()
```

```python
charactersToFilter2 = 'POA'
filteredRows2 = CarData[CarData['Price'].str.contains(charactersToFilter2)]
CarData = CarData.drop(filteredRows2.index)
```

#### Filtered Rows Based On Brand

```python
filteredBrand = CarData[CarData['Brand'].str.contains(charactersToFilter)]
```

```python
CarData = CarData.drop(filteredBrand.index)
```

```python
CarData.info()
```

```python
filteredColour = CarData[CarData['ColourExtInt'].str.contains(charactersToFilter)] 
```

```python
CarData = CarData.drop(filteredColour.index)
```

```python
CarData.info()
```

```python
FilterFuelConsumption = CarData[CarData['FuelConsumption'].str.contains(charactersToFilter)] 
```

```python
CarData = CarData.drop(FilterFuelConsumption.index)
```

```python
CarData.info()
```

```python
CarData['Kilometres'] = CarData['Kilometres'].astype('Int32')
```

```python
CarData.dtypes 
```

```python
charactersToRemove = ','
```

```python
#for columns in CarData.columns: 
 #   if CarData[columns].dtype == 'object':
 #       CarData[columns] = CarData[columns].str.replace(charactersToRemove, '') 
```

```python
 CarData['Engine'] = CarData['Engine'].str.replace(charactersToRemove, '') 
```

```python
CarData.head(10)
```

#### Removing the first 5 Characters in Engine column

```python
remove5Characters = 6; 
```

```python
CarData['Engine'] = CarData['Engine'].apply(lambda x: x[remove5Characters:])
```

```python
CarData.head() 
```

```python
CarData['Doors'] = CarData['Doors'].apply(lambda x: x[:-remove5Characters])
```

```python
CarData.head()
```

```python
CarData['Seats'] = CarData['Seats'].apply(lambda x: x[:-remove5Characters])
CarData.head()
```

```python
CarData['FuelConsumption'] = CarData['FuelConsumption'].apply(lambda x: x[:-10])
CarData.head()
```

```python
CarData['CylindersinEngine'] = CarData['CylindersinEngine'].apply(lambda x: x[:-4])
CarData.head()
```

```python
CarData['Engine'] = CarData['Engine'].apply(lambda x: x[:-1])
CarData.head()
```

```python
CarData['Doors'] = CarData['Doors'].astype('int32')
```

```python
CarData['Seats'] = CarData['Seats'].astype('int32')
```

```python
CarData.head(170)
```

```python
CarData['Engine'] = pd.to_numeric(CarData['Engine'])
```

```python
CarData['FuelConsumption'] = pd.to_numeric(CarData['FuelConsumption'])
```

```python
CarData['Price'] = pd.to_numeric(CarData['Price'])
```

```python
CarData['CylindersinEngine'] = pd.to_numeric(CarData['CylindersinEngine'])
```
